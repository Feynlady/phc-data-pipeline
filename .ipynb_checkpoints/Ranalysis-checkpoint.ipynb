```python
def append(data, run):
    result = pd.DataFrame()
    for a in run:
        f = subprocess.run("guild select " +a, shell=True, capture_output=True)
        d = pd.read_csv('venv/.guild/runs/'+f.stdout.decode('utf-8')[:-1]+'/scores.csv')
        d['run'] = [a for i in range(0, d.shape[0])]
        d = d.set_index('run')
        pd.concat(result, d)
    return data
```
